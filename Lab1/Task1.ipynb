```python
import matplotlib.pyplot as plt

# Creating the arrays
exponential_data = np.random.exponential(scale=1.0, size=100000)
uniform_data = np.random.rand(100000)
normal_data = np.random.randn(100000)

# Plotting the histograms
plt.hist(exponential_data, density=True, bins=100, histtype="step", color="blue", label="Exponential")
plt.hist(uniform_data, density=True, bins=100, histtype="step", color="green", label="Uniform")
plt.hist(normal_data, density=True, bins=100, histtype="step", color="red", label="Normal")

plt.legend(loc="upper right")
plt.title("Random distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```
